# Histogram and Barplots to show the comparison between countries

```python
df = pd.read_csv('/content/total_summary.csv')
df
```

```python
# Plot histgrams using seaborn
sns.set(style="whitegrid")

g = sns.histplot(df['Avg Happiness'], bins=20, kde=False)
plt.title('Distribution of Average Happiness Scores across Countries')
plt.show()
```

```python
g = sns.histplot(df['Avg GDP'], bins=20, kde=False)
plt.title('Distribution of Average GDP Scores across Countries')
plt.show()
```

```python
# Convert Df from wide form to long form
df_melted = df.melt(id_vars='Country', var_name='Metric', value_name='Value')
# df_melted
```

```python
g = sns.catplot(
    data=df_melted,
    x='Country',
    y='Value',
    hue='Metric',
    kind='bar',
    aspect=2,
    height=4
)
plt.title('Comparing Avg Happiness, GDP and Life Expectancy by Country');
```
